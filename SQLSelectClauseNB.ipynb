```python
df = spark.read.table("Dimcustomer")
display(df)
```

```python
df2 = df.where(df.YearlyIncome == 130000)
display(df2)

```

```python
df2 = df2.select(df2.CustomerKey, df2.NumberCarsOwned, df2.FirstName, df2.LastName, df2.DateFirstPurchase)
display(df2)
```

```python
from pyspark.sql.functions import *
df2 = df2.withColumn("Fullname", concat(df2.FirstName, lit(" "), df2.LastName))
display(df2)
```

```python
df2 = df2.withColumn("FormattedDate", date_format(df2.DateFirstPurchase,"EEEE d MMMM yyyy"))
df2 = df2.orderBy(df2.NumberCarsOwned.desc(), df2.CustomerKey)
display(df2)
```

```python
df3 = df.groupBy("EnglishOccupation").avg("YearlyIncome")
df3 = df3.withColumnRenamed("avg(YearlyIncome)","AverageIncome")
df3 = df3.withColumn("AverageIncome", df3.AverageIncome.cast("int"))
df3 = df3.where("AverageIncome > 70000")
# df3 = df3.sort(df3.AverageIncome.desc())
df3 = df3.orderBy(df3.AverageIncome, ascending=False)
display(df3)
```

```python
%%sql
SELECT CustomerKey, NumberCarsOwned, FirstName, LastName, DateFirstPurchase
FROM DimCustomer
WHERE YearlyIncome = 130000
```

```python
%%sql
SELECT CustomerKey, NumberCarsOwned,  DateFirstPurchase, concat(FirstName,' ',LastName) AS FullName,
date_format(DateFirstPurchase, "EEEE d MMMM yyyy") AS FormattedDate
FROM DimCustomer
WHERE YearlyIncome = 130000
ORDER BY NumberCarsOwned DESC, CustomerKey
```

```python
%%sql
SELECT EnglishOccupation, sum(TotalChildren) As Children , CAST(AVG(YearlyIncome) AS INT) AS AverageIncome

from DimCustomer
GROUP BY EnglishOccupation
HAVING CAST(AVG(YearlyIncome) AS INT) > 70000
ORDER BY Children DESC
```
