```python
df_test = spark.read.table("dimgeography")
display(df_test)
```

```python
#df_test2 = df_test.select("City", df_test.StateProvinceName.alias("State"))
df_test2 = df_test.select("City","StateProvinceName").withColumnRenamed("StateProvinceName", "State")
display(df_test2)
```

```python
df_test3 = df_test.where(df_test.GeographyKey >= 200)
display(df_test3)
```

```python
df_test4 = df_test3.withColumn("IPLocation", df_test3.IpAddressLocator.substr(2,3))
display(df_test4)
```

```python
#df_test5 = df_test3.where("CountryRegionCode = 'FR' or CountryRegionCode = 'GB'")
df_test5 = df_test3.where(df_test3.CountryRegionCode.isin("FR", "GB"))
display(df_test5)
```
