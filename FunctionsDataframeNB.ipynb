```python

df = spark.read.table("mtomactual")
display(df)
```

```python
from pyspark.sql.functions import *
df2 = df.withColumn("ColDate", add_months(lit("2026-01-01"), df.Actual/1000))
display(df2)
df2.write.mode("overwrite").format("delta").saveAsTable("MtoMActualWithDates")
```
